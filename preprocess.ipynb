```python
import datetime
datetime.datetime.strptime('2019-12-01 00:00:02 UTC','%Y-%m-%d %H:%M:%S UTC').date()
```

```python
def parsetime(x):
    try:
        return datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S UTC').date()
    except:
        return datetime.datetime.strptime('2019-12-01 00:00:02 UTC','%Y-%m-%d %H:%M:%S UTC').date()
```

```python
ddf = ddf[ddf.event_type == 'purchase']
ddf['event_time'].head()
```

```python
ddf['event_time'] = ddf['event_time'].apply(parsetime)
```

```python
ddf.head()
```

```python
a = ddf.groupby("event_time").price.sum().compute()
```
